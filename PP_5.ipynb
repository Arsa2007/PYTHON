```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

style.use("ggplot")
fb=(100,20,45,50,80,60,30)
yt=(110,25,23,56,20,20,70)
sn=(100,20,45,50,80,60,30)
xp = np.arange(len(fb))
plt.barh(xp,fb,color='blue',label='fb')
plt.barh(xp,yt,left=fb,color='red',label='yt')
plt.barh(xp,sn,left=yt,color='green',label='sn')
plt.xlabel('days')
plt.ylabel('time')
plt.title('time spend on social media')
plt.legend()
plt.show()

```
