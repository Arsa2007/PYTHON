```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

fb=(100,20,45,50,80,60,30)
yt=(110,25,23,56,20,20,70)
xp = np.arange(len(fb))

style.use("ggplot")
plt.figure(figsize=(7,5))
barwidth=0.2
plt.bar(xp,fb,color="blue",width=barwidth, label= 'fb')
plt.bar(xp+0.2,yt,color="red",width=barwidth, label= 'yt')
plt.xticks(xp+0.2, ('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'))
plt.xlabel('Days')
plt.ylabel("Time")

plt.title("Time spend on social media")
plt.legend()

```
